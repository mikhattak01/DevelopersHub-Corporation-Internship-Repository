## Data Visualization

Scatter plots to analyze relationships between given variables

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df, hue='Species')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=df, hue='Species')
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

Box plots to detect outliers and spread of values

```python
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Box Plot of Sepal Length by Species')
```

```python
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Box Plot of Sepal Width by Species')
```

```python
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Box Plot of Petal Length by Species')
```

```python
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Box Plot of Petal Width by Species')
```

## Summary

## 1. Sepal Characteristics:

**Setosa** has distinctly wider sepals but shorter sepal lengths compared to the other two species. **Versicolor** and **Virginica** overlap significantly in sepal dimensions, though Virginica tends to have slightly larger sepals.

**2. Petal Characteristics (Clearer Separation):**
**Setosa** has very small petals (both length and width), completely separated from the other two species. **Versicolor** has intermediate petal size. **Virginica** has the largest petals (longest and widest).

**3. Most Discriminating Feature:**
**Petal measurements** separate the three species much better than sepal measurements.

## Outliers

**No significant outliers** are visible in any of the box plots. The whiskers extend to typical ranges. This indicates:
The dataset is clean with no extreme values and each species shows relatively consistent measurements within its group. The box plots show compact distributions, especially for Setosa in petal dimensions

**The iris dataset appears well-behaved with clear species separation (particularly using petal features) and no anomalous measurements requiring removal.**
